### Exercise 2: Check Even or Odd

Write pseudocode that checks if a number is even or odd and prints the result. Then, write the equivalent Python code.

<!-- Add your pseudocode here -->

```python
# Pseudocode

# READ number
# IF number mod 2 == 0
# DISPLAY "even"
# ELSE DISPLAY "odd"

odd_or_even = int(input("Insert a number: "))

if odd_or_even % 2 == 0:
    print("The number is even")
else:
    print("The number is odd")
```

### Exercise 3: Sum of First N Natural Numbers

Write pseudocode to calculate the sum of the first N natural numbers. Then, write the equivalent Python code.

<!-- Add your pseudocode here -->

```python
# Pseudocode

# READ N
# DECLARE sum = 0
# ITERATE over natural numbers between 1 and N
# ADD all found numbers in variable sum
# DISPLAY sum

# Python

n = int(input("Insert a natural number: "))
sum = 0
for each in range(1,n+1):
    sum = sum + each
print(f'The sum of the first natural numbers until {n} is: {sum}')
```

### Exercise 4: Find Maximum of Three Numbers

Write pseudocode that finds the maximum of three numbers entered by the user. Then, write the equivalent Python code.

<!-- Add your pseudocode here -->

```python
# Pseudocode

# DECLARE num1, num2, num3, max
# READ users input and store in num1, num2, num3
# STORE max = num1
# IF num2 > max
# STORE max = num2
# IF num3 > max
# STORE max = num3
# DISPLAY max


num1 = int(input('Insert the first number: '))
num2 = int(input('Insert the second number: '))
num3 = int(input('Insert the third number: '))

highest = num1

if num2 > highest:
    highest = num2
if num3 > highest:
    highest = num3

print(f'The highest number is: {highest}')
```

### Exercise 5: Count Vowels in a String

Write pseudocode to count the number of vowels in a given string and print the count. Then, write the equivalent Python code.

<!-- Add your pseudocode here -->

```python
# PSEUDOCODE

# READ string
# DECLARE vowel_count = 0
# FOR each character in string
# IF character is a vowel
# increment vowel_count
# DISPLAY vowel_count

string = input('Enter a sentence:').lower()
vowel_count = 0
for each in string:
    if each in 'aeiou':
        vowel_count += 1
print(f'The vowel count of your string is {vowel_count}')
```

### Bonus Exercise 6: Reverse a String

Write pseudocode to reverse a given string and print it out.

<!-- Add your pseudocode here -->

```python
# pseudocode

# READ string
# DECLARE reversed_string = ''
# SET index = length of string - 1
# WHILE index >= 0
# add string[index] to reversed_string
# decrement index
# DISPLAY reversed_string

# python

string = input('Type a sentence for reversal: ')
reversed_string = ''
index = len(string)-1
while index >= 0:
    reversed_string += string[index]
    index -= 1
print(f'Your string in reverse: {reversed_string}')
```

### Bonus Exercise 7: Print Multiplication Table of a Number

Write pseudocode to print the multiplication table of a given number up to 10.

<!-- Add your pseudocode here -->

```python
# Pseudocode

# READ number
# FOR i from 1 to 10
# result = number * i
# DISPLAY number, 'x', i, '=', result

# python

number = int(input('Insert a number: '))
for i in range(1, 11):
    result = number * i
    print(number, 'x', i, '=', result)  
```

### Bonus Exercise 8: Count Words in a String

Write pseudocode that counts the words in a given string and prints the result.

<!-- Add your pseudocode here -->

```python
# Pseudocode

# READ string
# DECLARE count_words = 0
# DECLARE in_word = false
# FOR each char in string
    # IF character is not a space AND in_word is false
    #   increment count_words
    #   set in_word = true
    # ELSE IF char is a space
    #   set in_word = false
# DISPLAY count_words

string = input('Please introduce your string: ')
count_words = 0
in_word = False
for each in string:
    if each != ' ' and in_word == False:
        count_words += 1
        in_word = True
    elif each == ' ':
        in_word = False
print(f'Your word count is : {count_words}')
```
